```python
# Задано множину s, що складається з цілих чисел.
# Записати ВИРАЗ, значенням якого є множина, що складається з з елементів множини s, що діляться на 13.
var1 = { i for i in {13, 39, 26, 1, 2, 3} if not i%13 }

# Задано множину s, що складається з цілих чисел.
# Записати ВИРАЗ, значенням якого є множина, що складається з з непарних елементів множини s.
s = {1, 2, 3, 4, 5, 6, 7, 8, 9}
var2 = {i for i in s if i%2}

# Записати ВИРАЗ, значенням якого є множина,
# що складається з квадратних коренів цілих чисел від 15 до 2001 (включно), що не діляться на 2.
var3 = {i for i in range(15, 2002) if not (i**(1/2))%1 and i%2}

# Записати ВИРАЗ, значенням якого є словник, ключами якого є цілі числа від 14 до 2012 (включно),
# що діляться на 7, а ключам відповідають їх квадратні корені
var4 = {i: i*i for i in range(14, 2012) if not i%7}

# Записати ВИРАЗ, значенням якого є словник, ключами якого є цілі числа від 12 до 2000 (включно),
# що не діляться на 3, а ключам відповідають їх квадратні корені
var5 = {i: i*i for i in range(12, 2001) if i%3}

# Записати ВИРАЗ, значенням якого є словник, ключами якого є цілі числа від 11 до 2012 (включно),
# що діляться на 5, а ключам відповідають їх куби
var6 = {i: i*i for i in range(11, 2013) if not i%5}
```
